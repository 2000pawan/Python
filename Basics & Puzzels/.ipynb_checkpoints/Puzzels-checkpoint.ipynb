```python
def test(x):
    l=[]
    s=''
    for i in x.replace(' ', ''):
        s+=i
        if s.count("(")==s.count(")"):
            l.append(s)
            s=''
    return l
x='( ()) ((()()())) (()) ()'
print(test(x))
```

```python
def test(x):
    if " " in x:
        return x.split(' ')
    elif ',' in x:
        return x.split(',')
    else:
        return [c for c in x if c.islower() and ord(c)%2==0]
x='a,b,c,d'
print(test(x))
x='a b c d'
print(test(x))
x='abcd'
print(test(x))
```

```python
def test(x):
    if all(x[i]<x[i+1] for i in range(len(x)-1)):
        return 'Increasing'
    elif all(x[i+1]<x[i] for i in range(len(x)-1)):
        return 'Decreasing'
    else:
        return 'Not a monotonic sequence!'
x=[1, 2, 3, 4, 5, 6]
print(test(x))
x=[6, 5, 4, 3, 2, 1]
print(test(x))
x=[19, 19, 5, 5, 5, 5, 5]
print(test(x))
```

```python
def test(x):
    l=max([len(v) for v in x])
    for i in x:
        if len(i)==l:
            return i
x=['cat', 'catatatatctsa', 'abcdefhijklmnop', '124259239185125', '', 'foo', 'unique']
print(test(x))
x=['Green', 'Red', 'Orange', 'Yellow', '', 'White']
print(test(x))
```

```python
def test(x):
    s=set(x)
    for i in s:
        if -i in s:
            return [x.index(i), x.index(-i)]
x=[1, -4, 6, 7, 4]
print(test(x))
x=[1232, -20352, 12547, 12440, 741, 341, 525, 20352, 91, 20]
print(test(x))
```

```python
def test(x):
    return min(x,key=lambda i: sum(len(j) for j in i))
x=[['this', 'list', 'is', 'narrow'], ['I', 'am', 'shorter but wider']]
print(test(x))
x=[['Red', 'Black', 'Pink'], ['Green', 'Red', 'White']]
print(test(x))
```
