```python
def gcd(x,y):
    gcd=1
    if x%y==0:
        return y
    for k in range(int(y/2),0,-1):
        if x%k==0 and y%k==0:
            gcd=k
            break
    return gcd
print('GCD of 12 & 17=', gcd(12,17))
print('GCD of 4 & 6=', gcd(4,6))
print('GCD of 336 & 360=', gcd(336,360))
```

```python
def lcm(x,y):
    if x>y:
        z=x
    else:
        z=y
    while True:
        if (z%x==0) and (z%y==0):
            lcm=z
            break
        z+=1
    return lcm
print(lcm(4,6))
print(lcm(15,17))
```
