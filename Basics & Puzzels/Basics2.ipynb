```python
def remove_nums(l):
    p=2 # index of 3rd element is 2
    i=0
    len_l=len(l)
    while len_l>0:
        i=(p+i)%len_l
        print(l.pop(i))
        len_l-=1
l= [10, 20, 30, 40, 50, 60, 70, 80, 90]
remove_nums(l)
```
