```python
import datetime as dt
def addYears(d,years):
    try:
        return d.replace(year=d.year+years)
    except ValueError:
        return d+(dt.date(d.year+years,1,1)-(dt.date(d.year,1,1)))
print(addYears(dt.date(2023,1,1), -1))
print(addYears(dt.date(2023,1,1), 0))
print(addYears(dt.date(2023,1,1), 2))
print(addYears(dt.date(2000,2,29),1))
```

```python
import datetime as dt
def is_third_tuesday(x):
    d=dt.datetime.strptime(x,'%b %d, %Y')
    return d.weekday()==1 and 14< d.day<22
print(is_third_tuesday('Jun 23, 2015')) 
print(is_third_tuesday('Jun 16, 2015')) 
print(is_third_tuesday('Jul 21, 2015'))
##########################################################################
print('..................................................................')
from datetime import datetime, timedelta
def third_tuesday(year, month):
    first_day = datetime(year, month, 1)
    first_day_of_week = first_day.weekday()
    days_to_tuesday = (1 - first_day_of_week + 7) % 7
    first_tuesday = first_day + timedelta(days=days_to_tuesday)
    third_tuesday = first_tuesday + timedelta(days=14)
    return third_tuesday
year = 2023
month = 12 
result = third_tuesday(year, month)
print(f"The third Tuesday of {datetime(year, month, 1).strftime('%B %Y')} is on {result.strftime('%Y-%m-%d')}.")
```

```python
from datetime import timedelta, date
def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)
start_dt = date(2015, 12, 20)
end_dt = date(2016, 1, 11)
for dt in daterange(start_dt, end_dt):
    print(dt.strftime("%Y-%m-%d"))
```

```python
import time
def time_struct(s):
    print(' tm_year :', s.tm_year)
    print(' tm_mon :', s.tm_mon)
    print(' tm_mday :', s.tm_mday)
    print(' tm_hour :', s.tm_hour)
    print(' tm_min :', s.tm_min)
    print(' tm_sec :', s.tm_sec)
    print(' tm_wday :', s.tm_wday)
    print(' tm_yday :', s.tm_yday)
    print(' tm_isdst:', s.tm_isdst)
print('Local Time:-')
time_struct(time.localtime())
print('\nGmtime:-')
time_struct(time.gmtime())
```

```python
import time
import os
def zone_info():
    print('TZ  :', os.environ.get('TZ', '(not set)'))
    print('Timezone abbreviations: ',time.tzname)
    print('Timezone : {} ({})'.format(time.timezone, (time.timezone/36000)))
    print('DST timezone ', time.daylight)
    print('Time :', time.strftime('%X %X %Z'),'\n')
print('Default Zone:')
zone_info()
TIME_ZONES=[
    'Pacific/Auckland',
    'Europe/Berlin',
    'America/Detroit',
    'Singapore'
]
for zone in TIME_ZONES:
    os.environ["TZ"]=zone
    time.tzset()
    print(zone, ':')
    zone_info()
#Used .py file
```
