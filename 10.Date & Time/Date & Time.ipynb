```python
import datetime as dt
def all_sundays(year):
    x=dt.date(year,1,1)
    x+=dt.timedelta(6-x.weekday())
    while x.year==year:
        yield x
        x+=dt.timedelta(7)
for i in all_sundays(2023):
    print(i,end=',   ')
```

```python
import datetime as dt
def addYears(d,years):
    try:
        return d.replace(year=d.year+years)
    except ValueError:
        return d+(dt.date(d.year+years,1,1)-(dt.date(d.year,1,1)))
print(addYears(dt.date(2023,1,1), -1))
print(addYears(dt.date(2023,1,1), 0))
print(addYears(dt.date(2023,1,1), 2))
print(addYears(dt.date(2000,2,29),1))
```

```python
import datetime as dt
def is_third_tuesday(x):
    d=dt.datetime.strptime(x,'%b %d, %Y')
    return d.weekday()==1 and 14< d.day<22
print(is_third_tuesday('Jun 23, 2015')) 
print(is_third_tuesday('Jun 16, 2015')) 
print(is_third_tuesday('Jul 21, 2015'))
##########################################################################
print('..................................................................')
from datetime import datetime, timedelta
def third_tuesday(year, month):
    first_day = datetime(year, month, 1)
    first_day_of_week = first_day.weekday()
    days_to_tuesday = (1 - first_day_of_week + 7) % 7
    first_tuesday = first_day + timedelta(days=days_to_tuesday)
    third_tuesday = first_tuesday + timedelta(days=14)
    return third_tuesday
year = 2023
month = 12 
result = third_tuesday(year, month)
print(f"The third Tuesday of {datetime(year, month, 1).strftime('%B %Y')} is on {result.strftime('%Y-%m-%d')}.")
```

```python
import calendar
year=2023
month=12
print(calendar.monthrange(year,month)[1])
###########################################################################
print('......................................................................')
from datetime import datetime
def last_day_of_month(year, month):
    first_day_of_week, last_day = calendar.monthrange(year, month)
    return datetime(year, month, last_day)
year = 2023
month = 12
last_day = last_day_of_month(year, month)
print(f"The last day of {datetime(year, month, 1).strftime('%B %Y')} is a {last_day.strftime('%A')}, {last_day.strftime('%Y-%m-%d')}.")
```
