```python
def prime(x):
    for i in range(2,x-1):
        if x%i==0:
            print('Given Number',x,'is not Prime!')
        else:
            print('Given Number',x,'is Prime!')
x=int(input('Enter a number you want to check for prime or not!:-'))
prime(x)
```
